```python
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
df = pd.read_csv('Data_drop_year.csv')
```

```python
df.head()
```

```python
# Split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Comment'], df.drop(['Name','Time','Store','Comment','Aspect','polarity'],axis=1), test_size=0.2, random_state=42)
```

```python
# create a logistic regression model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# create a tfidf vectorizer object
vectorizer = CountVectorizer(max_df=0.8, max_features=5000)

# fit the vectorizer on the train set
vectorizer.fit(X_train)

# transform the train and test set
X_train_dtm = vectorizer.transform(X_train)
X_test_dtm = vectorizer.transform(X_test)
```

```python
%%time
# initialize classifier chains multi-label classifier
classifier = ClassifierChain(LogisticRegression())

# Training logistic regression model on train data
classifier.fit(X_train_dtm, y_train)

# predict
predictions = classifier.predict(X_test_dtm)

# accuracy
print("Accuracy = ",accuracy_score(y_test,predictions))
print("\n")
```

```python
# predict a comment
comment = ['The concept is great and the ai camera is so smart']

# remove punctuation
import string
comment = [char for char in comment if char not in string.punctuation]

# remove stopwords
from nltk.corpus import stopwords
stopwords.words('english')
comment = [word for word in comment if word not in stopwords.words('english')]
comment = ''.join(comment)

# lowercase
comment = comment.lower()

# print the predicted labels
print(classifier.predict(comment))
```

```python
# create a logistic regression model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, random_state=42, test_size=0.30, shuffle=True)

print(train.shape)
print(test.shape)

train_text = train['Comment']
test_text = test['Comment']
```

```python
train_text
```

```python
# create a tfidf vectorizer object
vectorizer = CountVectorizer(max_df=0.8, max_features=3000)

# transform the train and test set
X_train_dtm = vectorizer.fit_transform(train_text)
X_test_dtm = vectorizer.fit_transform(test_text)
```

```python
y_train = train.drop(labels = ['Comment','Name','Time','Store'], axis=1)

y_test = test.drop(labels = ['Comment','Name','Time','Store'], axis=1)
```

```python
#Using pipiple for applying Radom foresr and onevsrest classifier
pipeline = Pipeline([
    ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=1)),
])
```

```python
categories = list(df.columns.values)
categories = categories[4:]
print(categories)
```

```python
for category in categories:
    print('**Processing {} comments...**'.format(category))
    # train the model using X_dtm & y
    pipeline.fit(X_train_dtm, train[category])
    # compute the testing accuracy
    prediction = pipeline.predict(X_test_dtm)
    print('Test accuracy is {}'.format(accuracy_score(test[category], prediction)))
    print("\n")
```
