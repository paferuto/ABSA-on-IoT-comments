```python
import pandas as pd
```

```python
df = pd.read_csv('DataWithLabel.csv')
```

```python
df
```

```python
#drop unnecessary columns
df = df.drop(['Unnamed: 0','sentiment_score'], axis=1)
df
```

```python
#rename label as Aspect
df = df.rename(columns={'label':'Aspect'})
```

```python
df
```

```python
# convert aspect in each row to list
# remove [ and ] from aspect
df['Aspect'] = df['Aspect'].apply(lambda x: x.replace('[','').replace(']',''))
df['Aspect'] = df['Aspect'].apply(lambda x: x.split(','))
df
```

```python
# remove the first element in each row
df['Aspect'] = df['Aspect'].apply(lambda x: x[1:])
df
```

```python
# drop rows with empty aspect
df = df[df['Aspect'].map(len) > 0]
```

```python
df
```

```python
# save to csv
df.to_csv('Data_full.csv', index=False)
```

```python
#convert column time to datetime with format DD-MM-YYYY
df['Time'] = pd.to_datetime(df['Time'], format="%d-%m-%Y")
```

```python
df
```

```python
#drop rơ before 2018
df = df[df['Time'] > '2018-01-01']
```

```python
#save to csv
df.to_csv('Data_drop_year.csv', index=False)
```

```python
# remove rows with walmart as store
df = df[df['Store'] != 'Walmart']
```

```python
#save to csv
df.to_csv('Data_drop_walmart.csv', index=False)
```
