```python
import pandas as pd
```

```python
df = pd.read_csv('application.csv')
```

```python
df
```

```python
df['Application'] = df['Application'].astype(str)
```

```python
for i in range(len(df)):
    df['Application'][i] = df['Application'][i].split(', ')
df.head()
```

```python
# for each row in df, if column 'Application' have more than 1 value, then devide it into multiple rows with single value in 'Application' column
df2 = df.explode('Application')
df2
```

```python
df2.to_csv('application.csv', index=False)
```
