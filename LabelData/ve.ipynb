```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('application.csv')
```

```python
df.head()
```

```python
# show application unique values
df['Application'].unique()
```

```python
# plot bar chart showing distribution of polarity in each Application percentage with label on top of each bar
df.groupby('Application')['polarity'].value_counts(normalize=False).unstack().plot(kind='bar', stacked=False, figsize=(10,5),ylabel='Percentage', title='Distribution of polarity in each Application')
```
