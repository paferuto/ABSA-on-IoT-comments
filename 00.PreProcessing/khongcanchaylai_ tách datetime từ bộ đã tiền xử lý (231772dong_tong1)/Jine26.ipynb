```python
# keep number and remove character and alphabet in Time column
df['Time'] = df['Time'].str.replace('[a-zA-Z]', '')
df
```

```python
# 09 9 2022 to 09 09 2022 in Time column if len month is 1 digit and len day is 1 digit
df['Time'] = df['Time'].str.replace(' 9, ', ' 09 ')
df['Time'] = df['Time'].str.replace(' 8, ', ' 08 ')
df['Time'] = df['Time'].str.replace(' 7, ', ' 07 ')
df['Time'] = df['Time'].str.replace(' 6, ', ' 06 ')
df['Time'] = df['Time'].str.replace(' 5, ', ' 05 ')
df['Time'] = df['Time'].str.replace(' 4, ', ' 04 ')
df['Time'] = df['Time'].str.replace(' 3, ', ' 03 ')
df['Time'] = df['Time'].str.replace(' 2, ', ' 02 ')
df['Time'] = df['Time'].str.replace(' 1, ', ' 01 ')
df
```

```python
#remove , . space in Time column
df['Time'] = df['Time'].str.replace(',', '')
df['Time'] = df['Time'].str.replace('.', '')
#remove !@#$%^& punctuation in Time column
df['Time'] = df['Time'].str.replace('!', '')
df['Time'] = df['Time'].str.replace('@', '')
df['Time'] = df['Time'].str.replace('#', '')
df['Time'] = df['Time'].str.replace('$', '')
df['Time'] = df['Time'].str.replace('%', '')
df['Time'] = df['Time'].str.replace('^', '')
df['Time'] = df['Time'].str.replace('&', '')
df['Time'] = df['Time'].str.replace('*', '')
df['Time'] = df['Time'].str.replace('(', '')
df['Time'] = df['Time'].str.replace(')', '')
df['Time'] = df['Time'].str.replace('-', '')
df['Time'] = df['Time'].str.replace('_', '')
df['Time'] = df['Time'].str.replace('+', '')
df['Time'] = df['Time'].str.replace('=', '')
df['Time'] = df['Time'].str.replace('{', '')
df['Time'] = df['Time'].str.replace('}', '')
df['Time'] = df['Time'].str.replace('[', '')
df['Time'] = df['Time'].str.replace(']', '')
df['Time'] = df['Time'].str.replace('|', '')
df['Time'] = df['Time'].str.replace(':', '')
df['Time'] = df['Time'].str.replace(';', '')
df['Time'] = df['Time'].str.replace('"', '')
df['Time'] = df['Time'].str.replace("'", '')
df['Time'] = df['Time'].str.replace('<', '')
df['Time'] = df['Time'].str.replace('>', '')
df['Time'] = df['Time'].str.replace('?', '')
df['Time'] = df['Time'].str.replace('/', '')
df['Time'] = df['Time'].str.replace(' ', '')
df
```

```python
# remove space in Time column
df['Time'] = df['Time'].str.replace(' ', '')
#insert / in Time column at i =1 and i =3 
df['Time'] = df['Time'].str[:2] + '/' + df['Time'].str[2:4] + '/' + df['Time'].str[4:]
df
```

```python
#from string format 08 16 2008 to datetimw type 16/08/2008
df['Time'] = pd.to_datetime(df['Time'], format='%m/%d/%Y')
```

```python
#change  2022-09-22 to 22/09/2022
df['Time'] = df['Time'].dt.strftime('%d/%m/%Y')
df
```
