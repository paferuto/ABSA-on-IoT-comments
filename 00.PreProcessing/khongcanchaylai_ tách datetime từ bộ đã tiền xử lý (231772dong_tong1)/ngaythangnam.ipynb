```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df= pd.read_csv('thangngaynam.csv')
df
```

```python
df.dtypes
```

```python
#delete NaN values
df.dropna(inplace=True)
df = df.reset_index()
df
```

```python
# keep number and remove character and alphabet in Time column
df['Time'] = df['Time'].str.replace('[a-zA-Z]', '')
df
```

```python
#insert 0 in first if Time column len = 9
df['Time'] = df['Time'].apply(lambda x: '0' + x if len(x) == 9 else x)
df
```

```python
#insert 0 in first if Time column len = 9
df['Time'] = df['Time'].apply(lambda x: '0' + x if len(x) == 7 else x)
df
```

```python
#change if the end 22 to 2022 and 21 to 2021 and 20 to 2020 in Time column
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2022' if x[-3:] == '/22' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2021' if x[-3:] == '/21' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2020' if x[-3:] == '/20' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2019' if x[-3:] == '/19' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2018' if x[-3:] == '/18' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2017' if x[-3:] == '/17' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2016' if x[-3:] == '/16' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2015' if x[-3:] == '/15' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2014' if x[-3:] == '/14' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2013' if x[-3:] == '/13' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2012' if x[-3:] == '/12' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2011' if x[-3:] == '/11' else x)
df['Time'] = df['Time'].apply(lambda x: x[:-3] + '/2010' if x[-3:] == '/10' else x)

df
```

```python
#from string format 08 16 2008 to datetimw type 16/08/2008
df['Time'] = pd.to_datetime(df['Time'], format='%m/%d/%Y')
```

```python
#change  2022-09-22 to 22/09/2022
df['Time'] = df['Time'].dt.strftime('%d/%m/%Y')
df
```

```python
#to csv file
df.to_csv('thangngaynamx.csv', index=False)
```
